#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
# your code here

#Inicializo la figura
plt.figure(figsize=(12, 4))

#Para usar .sca debo hacer subplots
fig, ax = plt.subplots(1, 2) 

#Etiquetas de cada gráfico
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
dia = ['Weekend', 'Work']

#Inicio el primer gráfico
plt.sca(ax[0]) 

#Características
plt.title('Steps by months')
plt.xticks(range(1, 13), meses)

#Diseño del gráfico
plt.plot(fitbit.groupby('Months_encoded')['Steps'].mean(), c='steelblue', marker='o', linestyle='-')

#Inicio el segundo gráfico
plt.sca(ax[1]) 

#Características
plt.title('Steps en Work days or Weekend day')
colores = ['lightblue','mediumpurple']

#Para hacer un pie tengo que inicializarlo desde el axis
ax[1].pie(fitbit.groupby('Work_or_Weekend')['Steps'].mean(), labels=dia, autopct='%1.2f%%', startangle=90, colors=colores)

plt.savefig('../figures/1-Steps_by_months_and_weeddays.png')

plt.show();
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

plt.figure(figsize=(12, 4))

fig, ax = plt.subplots(1, 3) 

#Creo listas con nombres de las columnas a imprimir y los colores para cada una
graficos= ['Minutes Sedentary' , 'Minutes Lightly Active', 'Minutes Very Active']
colores = ['lightcoral','firebrick','darkred']


for i in range(3):
    plt.sca(ax[i])

    plt.scatter(x=fitbit[graficos[i]], y=fitbit['Steps'], c=colores[i])

    plt.title(graficos[i])

    plt.xlabel('Minutes')
    plt.ylabel('Steps')

plt.tight_layout()

plt.savefig('../figures/2-Scatter_Minutes_Steps.png')

plt.show();
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

# your code here

plt.figure(figsize=(12,6))

fig, ax = plt.subplots(1, 3) 

#Creo listas con nombres de las columnas a imprimir y los colores para cada una
bins_size= [10,20,50]
colores = ['lightcoral','firebrick','darkred']


for i in range(3):
    plt.sca(ax[i])

    plt.hist(x= titanic.Age, bins= bins_size[i])
    plt.title(f'Age with {bins_size[i]} bins')


    plt.xlabel('Ages')
    plt.ylabel('Passengers')


plt.tight_layout()
plt.savefig('../figures/4-Titanic_Ages_histS.png')

plt.show();
```

#### How does the bin size affect your plot?


A medida que los bins son más pequeños, se acercan más al tamaño real de la variable, 
es por esto que el eje vertical disminuye su rango a medida que va disminuyendo el tamaño de los bins

#### Plot the correlation matrix using seaborn.

```python
# your code here
sns.set(style='white')       # estilo blanco

mascara = np.triu(np.ones_like(titanic.corr(numeric_only=True), dtype=bool))   # mascara


sns.set(style='white', rc={"axes.edgecolor": "black"})  #Este es mi tercer intento de mostrar las líneas de los ejes y no funciona


sns.heatmap(titanic.corr(numeric_only=True), cmap='inferno',  mask=mascara,                 
            vmax=1, vmin=-1, center=0, linecolor='black', square=True, annot=True  
           )


plt.savefig('../figures/13-Titanic_corr_heatmap.png')

plt.show();
```
