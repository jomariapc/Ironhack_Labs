## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def uniform_random(bottom, ceiling, count):

    x = uniform.rvs(size=count)
    
    return bottom + (ceiling-bottom)*x    
```

```python
muestra_1= uniform_random(10,15,100)
muestra_2= uniform_random(10,60,1000)

plt.subplot(1, 2, 1)
sns.histplot(muestra_1, element='step', bins=10);
plt.title('Muestra 1')

plt.subplot(1, 2, 2)
sns.histplot(muestra_2, element='step', bins=10);
plt.title('Muestra 2')

plt.tight_layout()

plt.show()
```

How are the two distributions different?

The difference is that sample 1 is composed of 100 elements ranging from 10 to 15, and sample 2 is composed of 1000 elements ranging from 10 to 60.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here    
def normal_random(mean, std_dev, count):

    return norm.rvs(loc=mean, scale=std_dev, size=count)
```

```python
muestra_3= normal_random(10, 1, 1000)
muestra_4= normal_random(10, 50, 1000)

plt.subplot(1, 2, 1)
sns.histplot(muestra_3, element='step', bins=10);
plt.title('Muestra 3')

plt.subplot(1, 2, 2)
sns.histplot(muestra_4, element='step', bins=10);
plt.title('Muestra 4')

plt.tight_layout()

plt.show()
```

How are the two distributions different?

The standard deviation for each distribution is different (1 and 50), therefore it changes the distribution and the range of the x-axis.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def exp_random(count):

    return np.random.exponential(scale=10,  size=count)
```

```python
muestra_5= exp_random(10000)

muestra_6= exp_random(100000)


plt.subplot(1, 2, 1)
sns.histplot(muestra_5, element='step', bins=100);
plt.title('Muestra 5')

plt.subplot(1, 2, 2)
sns.histplot(muestra_6, element='step', bins=100);
plt.title('Muestra 6')

plt.tight_layout()

plt.show()
```

How are the two distributions different?

Sample 6 has more events, therefore the exponential is better distributed and reaches higher frequencies.
